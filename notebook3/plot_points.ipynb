## Simple Dot Plot _(Point Scater)_

```python
import matplotlib.pyplot as plt
```

```python
# x and y coordinates
x = [-4, -2, 0, 2, 4]
y = [3, 1, -1, -3, -5]
```

- Plot the dots

```python
plt.scatter(x, y)
plt.show()
```

- add `Title` to the graph

```python
plt.title('Simple Scatter Plot')
plt.scatter(x, y)
plt.show()
```

- add the label to the axes

```python
plt.xlabel("X Axe")
plt.ylabel("Y Axe")
plt.scatter(x, y)
plt.show()
```

- set the size and color of the dots

```python
plt.scatter(x, y, s=60, c='red')
plt.show()
```
